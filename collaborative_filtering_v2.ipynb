```python
# define X,W,b
tf.random.set_seed(1234)
len_recipe, users_count = Y.shape
recipe_x = tf.Variable(tf.random.normal(shape=(len_recipe,100)))
weight = tf.Variable(tf.random.normal(shape=(users_count, 100)))
bias  = tf.Variable(tf.random.normal(shape=(1, users_count)))
```

```python
def loss_function(X, W, b, Y, lambda_): 
    loss = (tf.linalg.matmul(X , tf.transpose(W)) + b - Y)
    loss = 0.5 * tf.reduce_sum(loss**2) + (lambda_/2) * (tf.reduce_sum(X**2) + tf.reduce_sum(W**2))
    return loss    
```

```python
# hyperparameter tuning
iteration = 2000
optimizer = tf.keras.optimizers.Adam()

for i in range(iteration): 
    with tf.GradientTape() as tape: 
        #  apply the loss function that haas been build before
        loss = loss_function(recipe_x, weight, bias, Y, lambda_ = 1)
        #  apply gradient tape processing to recipe_x, user_weight, bias variable
    grads = tape.gradient(loss, [recipe_x, weight, bias])

    # applyy recipe_x, weight, bias to optimizer
    optimizer.apply_gradients(zip(grads, [recipe_x, weight, bias]))
    # print the loss information 
    if i % 20 == 0: 
        print(f"Training loss at iteration {i}: {loss:0.1f}")
        # print(f'Training loss at iteration {i}: {loss}')
```

```python
# creating the model
# the user ratings will be used as validation dataset will
class customCallbacks(tf.keras.callbacks.Callback): 
    def on_train_end(self, logs={}):
        if logs.get('loss') <= 9.4887e-08: 
             self.model.stop_training = True

model = tf.keras.Sequential(
    layers = (
        tf.keras.layers.Dense(activation='linear', units = 5000, input_shape = (None,200)), 
        tf.keras.layers.Dense(activation='linear', units = 1000), 
        tf.keras.layers.Dense(activation='linear', units = 512), 
        tf.keras.layers.Dense(activation='linear', units = 128),
        tf.keras.layers.Dense(activation='linear', units = 64),
        tf.keras.layers.Dense(activation='linear', units =1)
    )
)
callbacks = customCallbacks()
model.compile(optimizer=tf.keras.optimizers.Adam(),loss='mean_squared_error', metrics='mse')
# x = (np.matmul(recipe_x.numpy(), np.transpose(weight)) + bias.numpy())[:, :175]
# validation = x[:, :-25]
history = model.fit(x=(np.matmul(recipe_x.numpy(), np.transpose(weight)) + bias.numpy()), y = Y, epochs = 5, callbacks=callbacks)
```
